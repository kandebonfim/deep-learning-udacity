```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_fwf("brain_body.txt")
```

```python
x = df[['Body']]
y = df[['Brain']]

plt.scatter(x, y)
plt.xlabel('Body')
plt.ylabel('Brain')
plt.show()
```

```python
model = linear_model.LinearRegression()
model.fit(x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r-')
plt.xlabel('Body')
plt.ylabel('Brain')
plt.show()
```
